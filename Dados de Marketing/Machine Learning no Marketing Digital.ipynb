# Feature Engineering

```python
# Sabemos que temos mais visitas do que usuários no site. Precisamos mostrar o que o usuário fez ao longo de sua jornada neste
#período que estamos analisando. Vamos poder identificar um usuário que tem algum tipo de comportamento e prever quanto ele 
#vai gastar.
#Precisamos ser capazes de calcular por visitorid o que aconteceu e os resultados desse usuário.
df_quant = df.groupby('fullVisitorId',as_index=False)[quant].sum()
```

# Melhorando o Feature Engineering

João é um tipo de usuário que fez mais de um acesso. Talvez seja interessante ver informação sobre a primeira e a última visita dele, para entender o primeiro device, o último. Se ele veio de uma mídia orgânica, se foi impactado. Esse tipo de informação costuma ser bem rica para alimentar nosso modelo. Vamos considerar para cada usuário os valores quantitativos e qualitativos da primeira e última visita.

A nossa estratégia para cada usuário será a seguinte: mostrar as informações qualitativas da primeira e última sessão, e a soma dos valores quantitativos. Como primeiro passo, vamos criar um Dataframe com o fullVisitorId e com a última sessão que temos para aquele usuário no site.

```python
visitas_ultima = df.groupby('fullVisitorId',as_index=False)
```

```python
visitas_ultima = visitas_ultima['visitNumber'].max()
```

```python
#Temos um novo dataframe com informações. Ele me diz o máximo de visita de cada usuário. 
visitas_ultima.head()
```

```python
df.head()
```

```python
usuarios_visitas_unicos = df.drop_duplicates(subset=['fullVisitorId','visitNumber'])
```

```python
usuarios_visitas_unicos.head()
```

```python
usuarios_visitas_unicos.shape
```

```python
visitas = pd.merge(visitas_ultima,usuarios_visitas_unicos,left_on=['fullVisitorId','visitNumber'],
                  right_on=['fullVisitorId','visitNumber'],how='left')
```

```python
visitas.head()
```

```python
visitas.shape
```

Assim como fizemos para a última sessão, vamos criar um DataFrame com o usuário e a sua primeira sessão no site:

```python
visitas_primeira = df.groupby('fullVisitorId',as_index=False)
```

```python
visitas_primeira = visitas_primeira['visitNumber'].min()
```

```python
visitas_primeira.head()
```

```python
visitas.head()
```

```python
visitas = pd.merge(visitas,usuarios_visitas_unicos,left_on=['fullVisitorId','visitNumber_primeira'],
                  right_on=['fullVisitorId','visitNumber'],how='left', suffixes=['_ultima','_primeira'])
```

Agora já temos um dataframe com todas as informações sobre primeira e última visita do usuário.
